# Data Cleaning and Transformation

```python
retail_data = pd.concat(
    [retail1, retail2],
    ignore_index=True
)

print("Combined Dataset Shape:")
print(retail_data.shape)
```

```python
print(
    retail_data["price"]
    .isnull()
    .sum()
)
```

```python
price_lookup = (
    products
    .set_index("product_id")["price"]
    .to_dict()
)
```

```python
retail_data["price"] = retail_data.apply(
    lambda row:
    price_lookup.get(
        row["product_id"],
        row["price"]
    )
    if pd.isna(row["price"])
    else row["price"],
    axis=1
)
```

```python
retail_data["price"].isnull().sum()
```

```python
retail_data["category"] = (
    retail_data["category"]
    .replace({
        "ELEC": "Electronics",
        "electronics": "Electronics",

        "FURN": "Furniture",
        "furniture": "Furniture",

        "HOME": "Home Appliances",
        "home appliances": "Home Appliances",

        "CLOTH": "Clothing",
        "clothing": "Clothing"
    })
)
```

```python
print(
    retail_data["category"]
    .unique()
)
```

```python
retail_data["transaction_date"] = pd.to_datetime(
    retail_data["transaction_date"],
    errors="coerce"
)
```

```python
retail_data["transaction_date"].head()
```

```python
(
retail_data["quantity"] < 0
).sum()
```

## Quantity Validation

A total of 65 records contained negative quantities.

Since retail sales cannot have negative quantities, these records were classified as invalid transactions and removed from the curated dataset.

```python
invalid_quantity_count = (
    retail_data["quantity"] < 0
).sum()

print("Invalid Records:", invalid_quantity_count)

retail_data = retail_data[
    retail_data["quantity"] > 0
]

print("New Shape:")
print(retail_data.shape)
```

## PII Protection

Personally Identifiable Information (PII) fields were masked to protect customer privacy.

Fields masked:
- customer_name
- email
- phone

```python
def mask_name(name):
    if pd.isna(name):
        return name

    return name[0] + "***"
```

```python
retail_data["customer_name"] = (
    retail_data["customer_name"]
    .apply(mask_name)
)
```

```python
def mask_email(email):
    if pd.isna(email):
        return email

    username, domain = email.split("@")

    return username[:2] + "***@" + domain
```

```python
retail_data["email"] = (
    retail_data["email"]
    .apply(mask_email)
)
```

```python
def mask_phone(phone):
    phone = str(phone)

    return "******" + phone[-4:]
```

```python
retail_data["phone"] = (
    retail_data["phone"]
    .apply(mask_phone)
)
```

```python
retail_data[
[
"customer_name",
"email",
"phone"
]
].head()
```

## Revenue Calculation

Revenue was calculated using:

Revenue = Quantity × Price × (1 - Discount)

```python
retail_data["revenue"] = (
    retail_data["quantity"]
    *
    retail_data["price"]
    *
    (1 - retail_data["discount"])
)
```

```python
retail_data[
[
"quantity",
"price",
"discount",
"revenue"
]
].head()
```

```python
retail_data["year"] = (
    retail_data["transaction_date"]
    .dt.year
)

retail_data["month"] = (
    retail_data["transaction_date"]
    .dt.month_name()
)

retail_data["quarter"] = (
    retail_data["transaction_date"]
    .dt.quarter
)
```

## Product Name Standardization

Product names contained inconsistent capitalization patterns.

Examples:
- Laptop, LAPTOP, laptop
- Sofa, SOFA, sofa
- Phone, PHONE, phone

To ensure accurate reporting and aggregation, product names were standardized using title case formatting.

```python
retail_data["product_name"] = (
    retail_data["product_name"]
    .str.strip()
    .str.title()
)
```

```python
top_products = (
    retail_data
    .groupby("product_name")["revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products
```
